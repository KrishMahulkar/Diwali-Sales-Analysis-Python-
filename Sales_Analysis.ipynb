# Answering business questions



Which products should be promoted during Diwali?

```python
category_sales = df.groupby("Product_Category")["Total_Sales"].sum().sort_values(ascending=False)
print("\nTop Product Categories:\n", category_sales)

category_sales.plot(kind="bar", figsize=(12,6), color='skyblue')
plt.title("Top 10 Product Categories by Sales (Diwali)")
plt.xlabel("Product Category")
plt.ylabel("Total Sales (INR)")
plt.xticks(rotation=45)
plt.show()
```

I analyzed Total Sales by Product Category and found that Food contributed the highest revenue (~83M INR), followed by Clothing & Apparel and Electronics. Lower-performing categories included Stationery, Auto, and Beauty. This insight can help businesses prioritize marketing spend, optimize inventory, and design targeted Diwali campaigns.
This insight can help businesses:


*   Prioritize inventory and promotions for Food, Clothing, Electronics, and Footwear during festive sales
*   Create attractive Diwali offers and bundles around high-demand categories






Identifying most profitable customer segment

```python
age_sales = df.groupby("Age Group")["Total_Sales"].sum().sort_values(ascending=False)
print("\nSales by Age Group:\n", age_sales)

age_sales.plot(kind="bar", figsize=(12,6), color='purple')
plt.title("Total Sales by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Total Sales (INR)")
plt.show()
```

I analyzed Total Sales by Age Group and found that customers aged 26–35 generated the highest revenue (over 100M INR), making them the most valuable customer segment. This was followed by the 36–45 and 18–25 age groups, showing strong purchasing power among young adults and working professionals.
This insight can help businesses:

*   Focus marketing campaigns on the 26–35 and 36–45 age groups for maximum ROI
*   Optimize inventory toward products preferred by high-revenue age groups




Which gender contributes more to festive spending?

```python
gender_sales = df.groupby("Gender")["Total_Sales"].sum()
print("\nSales by Gender:\n", gender_sales)

gender_sales.plot(kind="bar", color=['lightblue','pink'])
plt.title("Total Sales by Gender")
plt.xlabel("Gender")
plt.ylabel("Total Sales (INR)")
plt.show()
```

I analyzed Total Sales by Gender and found that female customers (F) contributed significantly higher revenue (around 185M INR) compared to male customers (M) (around 78M INR). This shows that females are the primary revenue-driving segment in this dataset.



ZONE-WISE PERFORMANCE

```python
zone_sales = df.groupby("Zone")["Total_Sales"].sum().sort_values(ascending=False)
print("\nSales by Zone:\n", zone_sales)

zone_sales.plot(kind="pie", figsize=(8,8), autopct='%1.1f%%', startangle=140)
plt.title("Sales Distribution by Zone")
plt.ylabel("")
plt.show()
```


I analyzed Sales Distribution by Zone and found that the Central zone
contributes the highest share of total sales ~39%, making it the strongest
performing region. This is followed by the Southern zone ~25.2% and the Western zone ~17.3%, which together form a major portion of overall revenue.
This insight can help businesses:

* Strengthen operations and customer engagement in the Central zone to maintain high performance.
* Expand marketing efforts in the Northern and Eastern zones to boost sales.
*  Design zone-specific promotional strategies to balance sales distribution   across regions.

Which states generate the most revenue during Diwali?

```python
state_sales = df.groupby("State")["Total_Sales"].sum().sort_values(ascending=False)
print("\nTop 10 States by Sales:\n", state_sales.head(10))

state_sales.head(10).plot(kind="bar", figsize=(12,6), color='salmon')
plt.title("Top 10 States by Total Sales (Diwali)")
plt.xlabel("State")
plt.ylabel("Total Sales (INR)")
plt.xticks(rotation=45)
plt.show()
```
